2)	Utilizando a imagem ctskull-256.tif, refaça a quantização de seus pixels utilizando de 7 a 1 bit(s) por pixel.  Fazer os cálculos por meio do seu algortimo. Não usar o opencv para a conversão

```python

def amostrando(original_dpi,sampling_dpi,image):
    image_array = np.array(image)

    # Calculate scaling factors for width and height
    original_dpi = 1250
    target_dpi = 300

    sampling_scale=original_dpi/sampling_dpi

    new_width = int(image_array.shape[1] / sampling_scale)
    new_height = int(image_array.shape[0] / sampling_scale)

    # Create a new downscaled image array using NumPy
    downsampled_array = np.zeros((new_height, new_width), dtype=np.uint8)

    # Perform the downsampling by selecting pixels from the original array
    for y in range(new_height):
        for x in range(new_width):
            downsampled_array[y, x] = image_array[int(y * sampling_scale), int(x * sampling_scale)]

    # Convert the downscaled array back to an image using PIL
    return(downsampled_array)
```

```python
x=amostrando(1250,300,img)
plt.imshow(x, cmap='gray', vmin=0, vmax=255);
```

```python
# Load the image using OpenCV
image_path = "ctskull-256.tif"
skull = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if skull is None:
    print("Error loading the image.")
else:
    # Requantize the image using NumPy
    num_bits = 2
    max_value = 2 ** num_bits - 1
    novo_skull = np.floor_divide(skull, 256 // max_value)



```
