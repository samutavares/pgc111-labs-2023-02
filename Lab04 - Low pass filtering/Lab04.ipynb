1)	Aplique o filtro de Média a imagem abaixo para diferentes tamanhos de máscaras.
- 3x3; 7x7; 3x3 aplicado três vezes sequenciais.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# code here

img1 = cv2.imread("sta2.png", cv2.IMREAD_GRAYSCALE)

plt.imshow(img1, cmap='gray', vmin=0, vmax=255);
```

```python
import cv2
import numpy as np

input_image = img1

kernel_size = 3  

def mean_filter(image, kernel_size):
    padding = kernel_size // 2
    filtered_image = np.copy(image)

    for i in range(padding, image.shape[0] - padding):
        for j in range(padding, image.shape[1] - padding):
            neighborhood = image[i - padding:i + padding + 1, j - padding:j + padding + 1]
            
            filtered_image[i, j] = np.mean(neighborhood)

    return filtered_image

filtered_image_3 = mean_filter(input_image, 3)
filtered_image_5 = mean_filter(input_image, 5)
filtered_image_7 = mean_filter(input_image, 7)
```

```python
plt.imshow(filtered_image_3, cmap='gray', vmin=0, vmax=255);
```

```python
plt.imshow(filtered_image_5, cmap='gray', vmin=0, vmax=255);
```

```python
plt.imshow(filtered_image_7, cmap='gray', vmin=0, vmax=255);
```

4)	Aplique o filtro de Mediana para a imagem abaixo, que apresenta um ruído salt and pepper
- 3×3 filter;	7×7 filter

```python
#  code here
img3 = cv2.imread("sta2noi1.png", cv2.IMREAD_GRAYSCALE)

plt.imshow(img3, cmap='gray', vmin=0, vmax=255);
```

```python
import cv2
import numpy as np


def median_filter(image, kernel_size):
    padding = kernel_size // 2
    filtered_image = np.copy(image)

    for i in range(padding, image.shape[0] - padding):
        for j in range(padding, image.shape[1] - padding):
            neighborhood = image[i - padding:i + padding + 1, j - padding:j + padding + 1]
            
            filtered_image[i, j] = np.median(neighborhood)

    return filtered_image

filtered_image_3 = median_filter(img3, 3)
filtered_image_5 = median_filter(img3, 5)
filtered_image_7 = median_filter(img3, 7)
```

```python
plt.imshow(filtered_image_3, cmap='gray', vmin=0, vmax=255);
```

```python
plt.imshow(filtered_image_5, cmap='gray', vmin=0, vmax=255);
```
